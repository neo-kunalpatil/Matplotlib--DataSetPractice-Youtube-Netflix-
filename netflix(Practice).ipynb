```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv("netflix_titles.csv")
# print(df.head())

df=df.dropna(subset=['type','release_year','rating','country','duration'])

type_countes= df['type'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(type_countes.index,type_countes.values,color=["orange","red"])
plt.title("number of movie Vs tv shows on netflix")
plt.xlabel("type")
plt.ylabel("countes")
plt.savefig("netflixcount.png")
plt.show()


rating_counts= df['rating'].value_counts()
plt.figure(figsize=(7,5))
plt.pie(rating_counts.values,labels=rating_counts.index,colors=["red","green"],autopct= "%1.1f%%")

plt.title("rating count distrbution")
plt.savefig("rating_counts.png")
plt.show()


year_counts= df["release_year"].value_counts()
plt.figure(figsize= (7,5))
plt.bar(year_counts.index, year_counts.values,color=["blue","green"])
plt.title("year wise movie realese counts")
plt.xlabel("year")
plt.ylabel("count")
plt.savefig("year_count.png")
plt.show()


year_count= df["release_year"].value_counts().sort_index()
plt.figure(figsize= (7,5))
plt.scatter(year_count.index, year_count.values,color="green")
plt.title("year wise movie realese counts")
plt.xlabel("year")
plt.ylabel("count")
plt.savefig("year_counts.png")
plt.show()




countary_counts=df["country"].value_counts().head(10)
plt.figure(figsize=(8,6))
plt.barh(countary_counts.index, countary_counts.values, color="teal",edgecolor="black")
plt.title("top 10 contys by number of shows")
plt.xlabel("number of shows")
plt.ylabel("contry")
plt.savefig("top_10_contry.png")
plt.show()


#subplots years = movie, tv show gorups

content_by_year= df.groupby(['release_year','type']).size().unstack().fillna(0)
# print(content_by_year)

fig,ax= plt.subplots(1,2,figsize=(12,5))
ax[0].plot(content_by_year.index, content_by_year['Movie'],color="blue")
ax[0].set_title("number of movies per year")
ax[0].set_xlabel("years ")
ax[0].set_ylabel("numer of movies")


ax[1].plot(content_by_year.index, content_by_year['TV Show'],color="orange")
ax[1].set_title("number of tv shows per year")
ax[1].set_xlabel("years ")
ax[1].set_ylabel("numer of Tv Shows")


fig.suptitle("comparison of movie and tv shows")
plt.savefig("subplots.png")
plt.tight_layout()
plt.show()






```
